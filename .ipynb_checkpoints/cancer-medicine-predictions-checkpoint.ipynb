```python
from Bio import Entrez
from Bio import Align
from Bio import SeqIO

Entrez.email="monjurmorshed793@gmail.com"
gene_data1 = Entrez.efetch(db="nucleotide", id="7157", rettype="fasta", retmode="fasta")
# print(gene_data1.read())
gene_data2 = Entrez.efetch(db="nucleotide", id="7040", rettype="fasta", retmode="fasta")
# print(gene_data1.read())

first_seq = SeqIO.parse(gene_data1, 'fasta')
second_seq = SeqIO.parse(gene_data2, 'fasta')

first_s = next(first_seq)
second_s = next(second_seq)

aligner =Align.PairwiseAligner()
aligner.mode='local'
alignments = aligner.align(first_s.seq, second_s.seq)

print(alignments.score)
```
